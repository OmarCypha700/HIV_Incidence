## HIV death rate

Reading file

```python
# load csv data for death rate
df3 = pd.read_csv(r'C:\Users\CYPHA\OneDrive\Documents\VisualizeGhana\death-rate-from-hivaids-who.csv')

df3.head()
```

Get data for Ghana only

```python
# select data for Ghana
df3 = df3.loc[df3['Entity'] == 'Ghana']
df3
```

Reset index to year clolumn

```python
# Reset index
df3 = df3.set_index('Year')
df3
```

Rename columns for clarity

```python
df3 = df3.rename(columns = {'Death rate per 100,000 people - Cause: HIV/AIDS - Sex: Both sexes - Age_group: ALLAges': 'Death_rate_per_100,000'})
df3
```

Drop unwanted columns

```python
death_rate = df3.drop(columns = {'Entity', 'Code'})
death_rate
```

## Visualization

First get the require range of data for the visualization

For Incidence of HIV in the past seven years

```python
incidence_viz = incidence.loc[2013:2019]
incidence_viz
```

Death rate of HIV in the past five years

```python
death_rate_viz = death_rate.loc[2013:2019]
death_rate_viz
```

## Plot a line graph

```python
# Creat subplots with one row and two columns
fig, axs = plt.subplots(1,2, figsize = (12,4))

sns.lineplot(x=incidence_viz.index, y=incidence_viz['Incidence_per_100,000'], marker='o', ax=axs[0])
axs[0].set_title('Incidence of HIV per 100,000')

sns.lineplot(x=death_rate_viz.index, y=death_rate_viz['Death_rate_per_100,000'], marker='o', ax=axs[1])
axs[1].set_title('HIV death rate per 100,000')
```
